```python
import torch
import numpy as np

# Aufgabe 1
a = torch.tensor(np.arange(0, 100).reshape(10, 10))
print(a)
```

```python
# Aufgabe 2
a_t = a.t()
print(a_t)
print('Same storage: ',id(a.storage()) == id(a_t.storage()))
```

```python
# Aufgabe 3
b = a_t[1:,1:]
print(b)
print('stride: ', b.stride())
print('size: ',b.size())
print('offset: ', b.storage_offset())
```

```python
# Aufgabe 4
x = torch.randn(10,10)
#print('x = ',x)
d = torch.det(x)
#print('Determinant = ',d)
if d != 0:
    inv = torch.inverse(x)
    #print('Inverse = ',inv)
    e = torch.round(torch.matmul(x,inv))
    print('Unit matrix = ',e)
```
